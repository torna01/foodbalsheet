Question 8: Perform a groupby operation on ‘Element’.  What year has the highest sum of Stock Variation?

```python
elements = data.groupby('Element')[['Y2014', 'Y2015', 'Y2016', 'Y2017']].sum()
total_variation = elements.sum()
max_year = total_variation.idxmax()
print(f'Year with highest sum of stock variation: {max_year}')
```

Question 10: What is the total number of unique countries in the dataset?

```python
total_unique_countries = data['Area'].nunique()
print(f'Total number of unique countries in the dataset: {total_unique_countries}')
```

Question 11: What is the total Protein supply quantity in Madagascar in 2015?

```python
madagascar = (data['Area'] == 'Madagascar') & (data['Element'] == 'Protein supply quantity (g/capita/day)')
madagascar_protein2015 = data.loc[madagascar, 'Y2015']
total_madagascar_protein2015 = madagascar_protein2015.sum()
print(total_madagascar_protein2015)
```

Question 12: Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the 7th lowest sum in 2017?

```python
columns = data[['Y2017', 'Area']]
grouped_columns = columns.groupby('Area')['Y2017'].sum()
sorted_sum= grouped_columns.sort_values()
seventhLS_2017 = sorted_sum.iloc[6]
area_7thLS2017 = sorted_sum[sorted_sum == seventhLS_2017].index
print(f'Area with the 7th lowest sum in 2017: {area_7thLS2017}')
```

Question 13: What is the total number and percentage of missing data in 2014 to 3 decimal places?

```python
data2014 = data[['Y2014']]
total_missing2014 = data2014.isnull().sum().sum()
total_values2014 = data2014.size
percentage_missing2014 = (total_missing2014 / total_values2014) * 100
print(f'Total number of missing data in 2014: {total_missing2014:.3f}')
print(f'Percentage of missing data in 2014: {percentage_missing2014:.3f} %')
```

Question 14: Perform a groupby operation on ‘Element’.  What is the total number of the sum of Processing in 2017?

```python
grouped_col = data.groupby('Element')['Y2017'].sum()
processing_sum2017 = grouped_col.get('Processing', 0)
print(f'Total sum of processing in 2017: {processing_sum2017}')
```

Question 17: Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the highest sum in 2017?

```python
selected_columns = data[['Y2017', 'Area']]
grouped_column = selected_columns.groupby('Area')['Y2017'].sum()
area_highest_sum_2017 = grouped_column.idxmax()
highest_sum_2017 = grouped_column.max()

print(f'area with the highest sum in 2017 is {area_highest_sum_2017} with the sum {highest_sum_2017}')
```

Question 20: What is the mean and standard deviation across the whole dataset for the year 2017 to 2 decimal places?

```python
data_2017 = data[['Y2017']]
mean_2017 = data_2017.mean().values[0]
std_dev_2017 = data_2017.std().values[0]
print(f'mean for the year 2017: {mean_2017:.2f}')
print(f'standard deviation for the year 2017: {std_dev_2017:.2f}')
```
